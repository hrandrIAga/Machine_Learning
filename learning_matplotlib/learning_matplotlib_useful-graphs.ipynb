#### **A. Scatter classification :**

* _the half of ML problems are **classification problems*_

```python
# Example of scatter classification with iris flower

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #load the dataset

iris = load_iris()

x = iris.data # the data from each image load in the array x
y = iris.target #{0, 1, 2} possible target class load in the array y

#ploat the 2 first column (dim 2) of x + one color per target class
plt.scatter(x[:,0], x[:,1], c=y, alpha=0.7, s=x[:,2]*50)  # alpha argument = transparency , c argument = size of the points
plt.xlabel('length of sepal')
plt.ylabel('widht of sepal')
```

#### **B. 3D graphs :**

* _modelize more than 2 variables_
* _OOD method to plot is required_

```python
from mpl_toolkits.mplot3d import Axes3D #import the 3D graphs

ax=plt.axes(projection='3d') 
ax.scatter(x[:,0],x[:,1],x[:,2], c=y)
```

```python
# 3d math function modelization

f = lambda x,y: np.sin(x) + np.cos(x+y)
X = np.linspace(0,5,100) # create the X axis
Y = np.linspace(0,5,100) # create the y axis
X, Y = np.meshgrid(X,Y) # create the mesh on each plan

Z=f(X,Y)

ax=plt.axes(projection='3d')

ax.plot_surface(X,Y,Z, cmap='plasma') #cmap argument = color 
```

#### **C. Histograms :**

* _modelize data distribution_


```python
# Example of scatter classification with iris flower

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris #load the dataset

iris = load_iris()

x = iris.data # the data from each image load in the array x
y = iris.target #{0, 1, 2} possible target class load in the array y

#modelize the distribution of x first column with an 1D histogram
plt.hist(x[:,0], bins = 20) #bins = number of section (10 if no argument)
plt.hist(x[:,1], bins = 20)
```

```python
#modelize the distribution of c column 1 and two with a 2d histogram
plt.hist2d(x[:,0],x[:,1])
plt.xlabel('length of sepal')
plt.ylabel('width of sepal')
plt.colorbar() #bar that associate color and distribution
```

```python
# to analyze an image : get the frequency of each color

from scipy import misc
face = misc.face(gray=True)
#plt.imshow(face,cmap = 'gray')
plt.hist(face.ravel(), bins=255)
plt.show()
```

#### **D. Contours graph :**

* _Modelize a 3d graph in 2D with a view from the top_
* _to see local extremums_

```python
plt.contourf(X, Y, Z, 20, cmap='plasma') # plt.contours(1darray on X axis, 1d array on Y axis, 1d array on z axis, k <number of lines ie. precision>)
plt.colorbar()
```

#### **E. Imshow :**
* _show all numpy array and matrix :_
> correlation matrix, mask for boolean indexing, optimization,...

```python
plt.imshow(np.corrcoef(x.T), cmap='plasma') #x.T because we want column correlations and not lines correlation
plt.colorbar()
```
